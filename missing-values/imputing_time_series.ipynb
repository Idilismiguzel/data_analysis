```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
df = pd.read_csv("./air-quality.csv", parse_dates=["Date"], index_col="Date")
```

```python
# Ffill method
# Replace nans with the last observed value

print(df["Ozone"][30:40])
df.fillna(method="ffill", inplace=True)
print(df["Ozone"][30:40])
```

```python
# Bfill method
# Replace nans with the next observed value

print(df["Ozone"][30:40])
df.fillna(method="bfill", inplace=True)
print(df["Ozone"][30:40])
```

```python
# .interpolate() model extends the sequence of values to the missing data

# Linear: Impute linearly or with equidistant values

print(df["Ozone"][30:40])
df.interpolate(method="linear", inplace=True)
print(df["Ozone"][30:40])
```

```python
# quadratic: Impute linearly or with equidistant values

print(df["Ozone"][30:40])
df.interpolate(method="quadratic", inplace=True)
print(df["Ozone"][30:40])
```

```python
# neareast: Impute linearly or with equidistant values

print(df["Ozone"][30:40])
df.interpolate(method="nearest", inplace=True)
print(df["Ozone"][30:40])
```

```python
# Visualizing the time series imputations
```

```python
df["Ozone"].plot(title="Ozone", marker="o", figsize=(30,10))
```

```python
# Ffill imputation
ffill_imp = df.fillna(method="ffill")
ffill_imp["Ozone"].plot(color="red", marker="o", linestyle="dotted", figsize=(30,10))
df["Ozone"].plot(title="Ozone", marker="o")
```

```python
# Bfill imputation
bfill_imp = df.fillna(method="bfill")
bfill_imp["Ozone"].plot(color="red", marker="o", linestyle="dotted", figsize=(30,10))
df["Ozone"].plot(title="Ozone", marker="o")
```

```python
# Linear Interpolation imputation
linear_int = df.interpolate(method="linear")
linear_int["Ozone"].plot(color="red", marker="o", linestyle="dotted", figsize=(30,10))
df["Ozone"].plot(title="Ozone", marker="o")
```

```python
# Quadratic Interpolation imputation
quadratic_int = df.interpolate(method="quadratic")
quadratic_int["Ozone"].plot(color="red", marker="o", linestyle="dotted", figsize=(30,10))
df["Ozone"].plot(title="Ozone", marker="o")
```

```python
# Quadratic Interpolation imputation
nearest_int = df.interpolate(method="nearest")
nearest_int["Ozone"].plot(color="red", marker="o", linestyle="dotted", figsize=(30,10))
df["Ozone"].plot(title="Ozone", marker="o")
```
