```python
import pandas as pd
import numpy as np
import missingno as msno
```

Possible reasons for missing data

* Values simply missing at random instances or intervals in a variable
* Values missing due to another variable
* Values missing due to the missingness of the same or another variable

Types of missingness
* Missing completely at random (MCAR)
* Missing at random (MAR)
* Missing not at random  (MNAR)

```python
df = pd.read_csv("./diabetes.csv")
```

```python
sorted_df = df.sort_values(by="Serum_Insulin")
msno.matrix(sorted_df) # You can see the pattern in missingness
```

```python
# You can use heatmap, correlation map and dendrogram to find the correlations between missingness
```

```python
# Missingness heatmap

msno.heatmap(df) 
```

```python
# skin_fold and serum_insulin are highly correlated in the missingness
# bmi and skin_fold has the lowest correlation
```

```python
# Missingness Dendrogram

msno.dendrogram(df)
```
