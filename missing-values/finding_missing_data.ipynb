```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
df = pd.read_csv("./air-quality.csv", parse_dates=["Date"], index_col="Date")
```

```python
# Number of missin values
df.isnull().sum()
```

```python
# Percentage of missing values
df.isnull().mean()*100
```

```python
# Missingno package
msno.bar(df)
```

```python
msno.matrix(df)
```

```python
msno.matrix(df, freq="M") # it is time series data you can set the frequency to "M"
```
